### [Solve](http://docs.sympy.org/latest/tutorial/solvers.html)

```python
from sympy import solve
import sympy
x= sympy.Symbol('x')
y= sympy.Symbol('y')
sympy.init_printing()


solve(sympy.sin(x), x)  
# solve(x + x**2 - 12, x)  # solve the equation for x asuming that x is a real number
```

```python
f_xy = x**2 - y  # define a function of two variables
```

```python
solve(f_xy, y)  # solve the equation for y
```

```python
solve(f_xy, x)  # solve the equation for x
```

```python
f_xy.subs([(y, sympy.symbols('k'))])  # y = k
# f_xy.subs([(y, 3), (x, 2)])  # y = 3, x = 2
# f_xy
```

### Good Practices

SymPy will solve an equation in all the possible ways. So if the equation has more than one solution, it will return all the solutions. If the equation has no solution, it will return an empty list.

`symbols`
- `real`: only real solutions
- `positive`: only positive solutions
- `negative`: only negative solutions
- `integer`: only integer solutions
- `rational`: only rational solutions
- `nonnegative`: only nonnegative solutions
- `nonpositive`: only nonpositive solutions
- `finite`: only finite solutions
- `infinite`: only infinite solutions
- `complex`: only complex solutions
- `extended_real`: only extended real solutions

```python
x = sympy.symbols('x')  # x is any number
print(solve(x**2+1, x))  # solve the equation for x asuming that x is a real number

x = sympy.symbols('x', real=True)  # r is a real number
solve(x**2+1, x)  # solve the equation for x asuming that x is a real number
```

### [Lambdify](http://docs.sympy.org/latest/modules/utilities/lambdify.html)

Lambdify is a function that **converts SymPy expressions into Python functions**. It is useful when you want to use SymPy to do symbolic computations but need to evaluate the expressions numerically.

```python
from sympy import lambdify

f = x**2 + 2*x + 1
f
```

```python
f_lambdified = lambdify([x], f)  # lambdify([symbols], function)
for i in range(10):  # evaluate the function for i = 0:9
    print(f_lambdified(i), end=' ')
```
