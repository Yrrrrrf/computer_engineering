# Subplot
Subplot allow us to plot multiple plots in one figure. We can specify the number of rows and columns and the plot number.

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
# Data for plotting
t = np.arange(0.0, 2.0, 0.01)  # 0.01 is the step size
s1 = np.sin(2 * np.pi * t)  # 2 * np.pi is the frequency
s2 = 1.2 * np.sin(4 * np.pi * t)  # 4 * np.pi is the frequency

fig, (ax1, ax2) = plt.subplots(2, 1)  # 2 rows, 1 column
ax1.plot(t, s1, 'g')
ax1.set(xlabel='time (s)', ylabel='voltage (mV)')
ax1.grid()

ax2.plot(t, 2 * s1, 'b--')
ax2.set(xlabel='time (s)', ylabel='some other voltage (mV)')
ax2.grid()
```

By default if a figure has only one subplot, it takes the whole figure. We can specify the subplot using the function subplot(). It has three parameters: `nrows`, `ncols` and `plot_number`. These three parameters are used to identify the location of the subplot in the figure.

```python
# Subplot with 2 rows and 2 columns
fig2, (x1, x2, x3) = plt.subplots(1, 3)  # 3 columns
x1.plot(t, s1, 'g')
x2.hist(s1, 10)
x3.scatter(t, s1, c='r', marker='*') # type: ignore  # c is color, marker is the shape
```
