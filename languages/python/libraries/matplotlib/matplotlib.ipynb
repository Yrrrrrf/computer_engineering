# Matplotlib
Is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms. Matplotlib can be used in Python scripts, the Python and IPython shell (ala MATLAB or Mathematica), web application servers, and six graphical user interface toolkits.


```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
import numpy as np
# linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)
np.linspace(0, 1, 11)  # Create a list of evenly-spaced numbers over a specified interval
```

```python
# Plot the list
# plot.plot(x, y, format_string, **kwargs)
# plt.plot(np.linspace(0, 1, 11))
plt.plot(np.linspace(0, 1, 11), np.linspace(0, 1, 11)**2, 'mx')  # Plot y = x^2

# Colors and line styles
# https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.plot.html
# r = red, b = blue, g = green, c = cyan, m = magenta, y = yellow, k = black, w = white
# - = solid line, -- = dashed line, -. = dash-dot line, : = dotted line
# o = circle, s = square, ^ = triangle, * = star, + = plus, x = x, D = diamond
```

```python
# Is possible to add a legend, title, axis labels, etc to the plot
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot Title')
plt.legend(['y = x^2'])
# plt.plot(np.linspace(0, 1, 11), np.linspace(0, 1, 11)**2, 'mx')
plt.plot(np.linspace(0, 1, 11), np.linspace(0, 1, 11)**3, 'g^')
```
