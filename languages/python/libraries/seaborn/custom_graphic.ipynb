# [Custom Graphic](https://seaborn.pydata.org/tutorial/aesthetics.html#customizing-with-matplotlib-parameters)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")  # Load the example tips dataset
print(tips.head())  # Print the first 5 rows of the dataset
sns.color_palette("hls", 8)  # Set color palette
```

```python
def sinplot(n=10, flip=1):
    '''Plot n sin curves with different phases and amplitudes'''
    x = np.linspace(0, 2*np.pi, 100)  # 100 points between 0 and 14
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)
```

### [Set style](https://seaborn.pydata.org/tutorial/aesthetics.html#seaborn-figure-aesthetics)

Controlling the style of the plots is done through the `set_style()` function.

```python
sns.set_theme(style="dark", palette="pastel", color_codes=True)  # Set the style
# style = "darkgrid", "whitegrid", "dark", "white", "ticks"  # grid lines
# palette = "deep", "muted", "pastel", "bright", "dark", "colorblind"  # or a list of colors
# context = "paper", "notebook", "talk", "poster"  # font size
# font = "serif", "sans-serif", "monospace"  # font family
# font_scale = 1, 2, 3, ...  # font size
# color_codes = True, False  # whether to use color codes for colors
sinplot()  # This is the plot
```

```python
sns.set_theme()  # Reset the style
sinplot()  # Plot again
```

### [Figure Aesthetics](https://seaborn.pydata.org/tutorial/aesthetics.html)

Drawing attractive figures is important. When making figures for yourself, as you explore a dataset, its nice to have plots that are pleasant to look at. Visualizations are also central to communicating quantitative insights to an audience, and in that setting its even more necessary to have figures that catch the attention and draw a viewer in.

Matplotlib is highly customizable, but it can be hard to know what settings to tweak to achieve an attractive plot. Seaborn comes with a number of customized themes and a high-level interface for controlling the look of matplotlib figures.

```python
sns.displot(data=tips, x="total_bill", kde=True, rug=True, bins=12, hue="sex")  # Plot the distribution of the total_bill column
    # rug -> small linethat represents the distribution
    # kde -> Kernel Density Estimation (smooth line)
    # hue -> if you want to separate the data by a column
    # bins -> number of bins
    # color -> red, blue, green, etc.
    # leyend -> True or False

plt.show()  # Show the plot
```

```python
sns.displot(data=tips, x="total_bill", y="tip")  # kind="hist" by default
sns.displot(data=tips, x="total_bill", y="tip", kind="kde")
```

Here we will use the **seaborn** library to create a graphic that shows the relationship between the **Total bill** and **Tip** columns.

```python
sns.jointplot(x="total_bill", y="tip", data=tips)  # kind="scatter" by default
    # kind -> "scatter", "kde", "hex", "reg", "resid"
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex")
```
