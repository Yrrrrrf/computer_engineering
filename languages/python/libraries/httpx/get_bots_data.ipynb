# Get Bots Data

Using [httpx](httpx.md) module, we can get the data from any website.

```python
import httpx  # allows to make http requests
```

```python
async def get_bots_data(url: str) -> str:
    '''
    Get the robots.txt file from a website and return it as a string
    '''
    async with httpx.AsyncClient() as client:  # create an async client
        response = await client.get(url)  # get the robots.txt file
        return response.text  # return the robots.txt file as a string
```

```python
url = 'https://choosealicense.com/licenses/mit/'


async def main():
    '''
    Get the robots.txt file from a website and print it
    '''
    robots_txt = await get_bots_data(url)  # get the robots.txt file
    print(robots_txt)  # print the robots.txt file


if __name__ == '__main__':
    # asyncio.run(main())  # run the main function
    await main()  # run the main function
```
