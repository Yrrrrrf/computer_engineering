### Plotting (with Matplotlib)

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
```

### Plotting a function

```python
x = np.linspace(0, 2*np.pi, 128)  # n points from 0 to 2pi
f = np.sin(x)  # f(x) = sin(x)
plt.plot(x, f)
```


### Plotting a 2D array

##### [Mesh grid](https://numpy.org/doc/stable/reference/generated/numpy.meshgrid.html#numpy.meshgrid)

```python
# where x^y are the domain and range
x = np.arange(0, 5)
y = np.arange(0, 4)
xv, yv= np.meshgrid(x, y)  # returns coordinate matrices from coordinate vectors
print(xv, xv.shape)
print(yv, yv.shape)
```

```python
# this will perform element-wise multiplication
f = xv**2 + yv**2  # f(x, y) = x^2 + y^2
f
```

```python
# plt.contourf(x, y, z, levels, colors, linestyles, linewidths, antialiased)
plt.contourf(xv, yv, f, levels=32)  # show the contour lines
# plt.pcolormesh(xv, yv, f)  # show the color of each pixel
plt.colorbar()
```
